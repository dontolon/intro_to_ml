## Cos'è un Notebook?

Un notebook è un documento interattivo composto da **celle**. Esistono due tipi di celle:

| Tipo | Cosa contiene | Come appare |
|---|---|---|
| **Cella di codice** | Codice Python eseguibile | Sfondo grigio, numero `[ ]` a sinistra |
| **Cella di testo** | Spiegazioni scritte in Markdown | Testo formattato, nessun numero |

Quando esegui una cella di codice, Python elabora il contenuto e mostra l'output **subito sotto** la cella stessa. Le celle di testo come questa non producono output: si "renderizzano" e mostrano solo testo formattato.

**Google Colab** è un servizio gratuito di Google che ti permette di eseguire notebook nel browser, senza installare nulla sul tuo computer. Tutto gira su una macchina virtuale di Google.

## 3.2 Normalizzazione delle feature GPS

Prima di passare le coordinate GPS alla rete, le normalizziamo nell'intervallo $[0, 1]$ con il **MinMaxScaler**.

### Perché normalizzare?

Le coordinate grezze hanno scale numeriche diverse:

| Feature | Esempio di valore grezzo |
|---|---|
| Latitude | 33.420 – 33.435 |
| Longitude | −111.940 – −111.920 |

Se usiamo questi valori direttamente, la rete vede differenze numeriche che non riflettono le distanze geografiche reali. La normalizzazione porta entrambe le coordinate sullo stesso piano.

### Formula del MinMaxScaler

$$x_{\text{norm}} = \frac{x - x_{\min}}{x_{\max} - x_{\min}}$$

- Il valore minimo nel training diventa **0**.
- Il valore massimo nel training diventa **1**.
- Tutti gli altri valori si trovano proporzionalmente in mezzo.

### Regola critica: fit solo sul training

Il `MinMaxScaler` calcola min e max **solo dal model-train**. Poi applica gli stessi parametri a validation e test. Se usassimo i dati di validation o test per calcolare min/max, "guarderemmo il futuro" — questo è **data leakage** e renderebbe la valutazione troppo ottimista.

````python
# Il MinMaxScaler si addestra (fit) SOLO sul model-train.
# Validation e test vengono trasformati con gli stessi parametri → nessun leakage.
gps_scaler = MinMaxScaler()
X_model_train = gps_scaler.fit_transform(X_model_train)
X_val         = gps_scaler.transform(X_val)
X_test        = gps_scaler.transform(X_test)

print("GPS features dopo la normalizzazione MinMax:")
print(f"  X_model_train — min per feature: {X_model_train.min(axis=0).round(4)}, "
      f"max per feature: {X_model_train.max(axis=0).round(4)}")
print(f"  X_val         — min per feature: {X_val.min(axis=0).round(4)}, "
      f"max per feature: {X_val.max(axis=0).round(4)}")
print(f"  X_test        — min per feature: {X_test.min(axis=0).round(4)}, "
      f"max per feature: {X_test.max(axis=0).round(4)}")
````

# 4. Metriche

Per valutare i modelli usiamo due metriche complementari: la **Top-K accuracy** (misura ML) e la **power loss** (misura fisica del link).

---

## 4.1 Top-K Accuracy

$$A_{\text{top-}k} = \frac{1}{N} \sum_{i=1}^{N} \mathbb{1}\!\left\{ m^{*}_{i} \in \hat{\mathbf{m}}_{i} \right\}$$

dove $N$ è il numero totale di campioni. Il termine $m^{*}_{i}$ è l'indice del beam ottimale per l'$i$-esimo campione (cioè il beam con massima potenza ricevuta), e $\hat{\mathbf{m}}_{i}$ è il vettore dei $k$ beam predetti con i punteggi più alti. La funzione $\mathbb{1}\{\cdot\}$ vale $1$ se la condizione è vera, $0$ altrimenti.

**Intuizione**: per $k=1$ misuriamo quante volte il beam predetto è esattamente quello giusto. Per $k=3$, contiamo anche quando il beam giusto è nella top-3. Più $k$ è grande, più è facile "centrare" — ma più beam dobbiamo effettivamente misurare.

---

## 4.2 Power Loss

Anche un beam quasi ottimale può essere accettabile, se la perdita di potenza rispetto al beam migliore è piccola. La power loss per un singolo beam predetto $m$ è:

$$\mathsf{PL}[\text{dB}] = 10 \log_{10} \!\left( \frac{y_{m^{*}}}{y_{m}} \right)$$

dove $y_{m^{*}}$ è la potenza del beam ottimale e $y_{m}$ è la potenza del beam predetto. Un valore vicino a **0 dB** indica nessuna perdita; valori superiori a **3 dB** sono generalmente considerati significativi.

Quando forniamo una lista di $k$ beam candidati $\hat{\mathbf{m}}$, usiamo il migliore tra essi:

$$\overline{\mathsf{PL}}[\text{dB}] = 10 \log_{10} \!\left( \frac{y_{m^{*}}}{\max_{i \in \hat{\mathbf{m}}} \, y_{i}} \right)$$

In pratica, le potenze misurate contengono rumore. Quando il denominatore è molto piccolo si aggiunge $\varepsilon = 10^{-12}$ per evitare divisioni per zero.

---

**Cosa non fare:** scegliere il modello solo dalla Top-1 accuracy se l'obiettivo reale è la qualità del link radio.

# 5. Baselines: prima dei modelli veri

Una baseline è un modello semplice che stabilisce un riferimento.

Se la Neural Network non batte una baseline stupida, abbiamo un problema. Le baseline aiutano anche a interpretare numeri come "50% accuracy": buono o cattivo rispetto a cosa?

Useremo due baseline:

1. **Most frequent Top-K**: suggerisce sempre i beam più frequenti nel training set.
2. **Random Top-K**: suggerisce beam casuali.

Le valutiamo sul **test set** per il confronto finale della sezione 8.

````python
# Baseline 1: beam più frequenti nel training set.
frequent_beams = pd.Series(y_train).value_counts().index.to_numpy()
frequent_top10 = frequent_beams[:10]

# Test set
frequent_topk = np.tile(frequent_top10, (len(y_test), 1))
random_topk   = np.vstack([np.random.default_rng(42).permutation(64)[:10] for _ in range(len(y_test))])

# Validation set (usato nella Zona Studenti per il confronto con le baselines)
frequent_topk_val = np.tile(frequent_top10, (len(y_val), 1))
random_topk_val   = np.vstack([np.random.default_rng(42).permutation(64)[:10] for _ in range(len(y_val))])
````

````python
baseline_topk = {
    "Most frequent": topk_accuracy(y_test, frequent_topk, max_k=10),
    "Random":        topk_accuracy(y_test, random_topk,   max_k=10),
}

baseline_loss = {
    "Most frequent": mean_power_loss_db(power_test, frequent_topk, max_k=10),
    "Random":        mean_power_loss_db(power_test, random_topk,   max_k=10),
}
````

# 7. Neural Network in PyTorch

Una **Neural Network** impara parametri interni — pesi numerici — che trasformano le coordinate GPS normalizzate in 64 score, uno per ogni possibile beam.

## Il ciclo di addestramento

Per ogni gruppo di esempi (batch), la rete esegue quattro passi:

1. **Forward pass** — il GPS entra nella rete, passa attraverso due layer con funzione ReLU, ed escono 64 score (chiamati logit).
2. **Loss (Cross-Entropy)** — confrontiamo i 64 score con il beam corretto. Se il beam giusto ha il punteggio più alto, la loss è bassa. L'obiettivo è minimizzarla.
3. **Backward pass** — la rete calcola quanto cambiare ogni peso per ridurre la loss (regola della catena / backpropagation).
4. **Optimizer step** — i pesi vengono aggiornati di un piccolo passo nella direzione giusta.

Si ripete per ogni **epoch** (un passaggio completo sul dataset).

## I quattro parametri che controllerai

| Parametro | Cosa regola | Troppo piccolo | Troppo grande |
|---|---|---|---|
| `hidden_dim` | Numero di neuroni nel layer nascosto (capacità) | Non impara schemi complessi → underfitting | Memorizza il train → overfitting |
| `learning_rate` | Dimensione del passo di aggiornamento dei pesi | Training lentissimo | Loss instabile, oscilla o diverge |
| `epochs` | Quante volte la rete rivede tutto il dataset | Non ha avuto tempo → underfitting | Su dataset piccoli → overfitting |
| `batch_size` | Quanti esempi per ogni aggiornamento | Aggiornamenti frequenti ma rumorosi | Aggiornamenti stabili ma meno frequenti |

## Segnali di allarme nelle training curves

- **Train loss scende, validation loss sale** → overfitting: il modello memorizza il training
- **Entrambe le loss rimangono alte** → underfitting: il modello non impara abbastanza
- **Loss che oscilla molto** → learning rate troppo alto
- **Loss che scende lentissimo** → learning rate troppo basso

## 9.4 Perché la latitude è più importante? Analisi statistica

I risultati dell’ablazione (sezione 9.3) mostrano che la rete addestrata usando solo la *latitude* ottiene prestazioni migliori rispetto a quella che usa solo la *longitude*.  
Questo suggerisce che la latitude contenga più informazione utile per distinguere i beam. Ma perché succede?

Per capirlo utilizziamo due misure statistiche complementari.

---

### Mutual Information $I(X;Y)$

La **Mutual Information** misura quanta informazione una variabile contiene su un’altra, anche quando la relazione non è lineare.

$$
I(X;Y)=\sum_{x,y} p(x,y)\log\frac{p(x,y)}{p(x)p(y)}
$$

Un valore alto significa che conoscere $X$ riduce molto l’incertezza su $Y$.

Nel nostro caso:

- $X$ = coordinata GPS (latitude oppure longitude)
- $Y$ = beam migliore

Quindi, se la Mutual Information tra *latitude* e beam è alta, significa che osservando la latitude possiamo prevedere abbastanza bene quale beam sarà selezionato.

#### Esempio intuitivo

Immaginiamo una strada lunga in direzione nord-sud:

- nella parte nord viene quasi sempre usato il beam 3,
- nella parte sud viene quasi sempre usato il beam 12.

In questo caso la *latitude* identifica chiaramente il beam, quindi la Mutual Information è alta.

Al contrario, se lungo la stessa latitude possono comparire molti beam diversi in modo casuale, allora conoscere la latitude aiuta poco e la Mutual Information sarà bassa.

La cosa importante è che questa misura funziona anche quando la relazione è complessa o non lineare, cosa molto comune nei canali wireless reali.

---

### Eta-quadro $\eta^2$ (ANOVA)

L’**Eta-quadro** misura quanta parte della variabilità di una feature è spiegata dalle classi beam.

$$
\eta^2=\frac{SS_{between}}{SS_{total}}
=
\frac{\sum_k n_k(\bar{x}_k-\bar{x})^2}
{\sum_i(x_i-\bar{x})^2}
$$

In pratica, confronta:

- quanto le medie delle classi beam sono separate tra loro,
- rispetto alla variabilità totale dei dati.

Un valore alto di $\eta^2$ significa che i beam occupano regioni differenti lungo quella coordinata.

#### Esempio intuitivo

Supponiamo di osservare la latitude di utenti associati a tre beam:

| Beam | Latitude media |
|---|---|
| Beam 1 | 10 |
| Beam 2 | 20 |
| Beam 3 | 30 |

Le classi sono ben separate: ogni beam occupa una zona diversa.  
In questo caso $\eta^2$ è alto, perché la coordinata latitude spiega bene quale beam viene usato.

Viceversa, se tutti i beam hanno latitude distribuite più o meno tra 10 e 30 senza separazione chiara, allora $\eta^2$ sarà basso: la coordinata non aiuta molto la classificazione.

---

In sintesi:

- la **Mutual Information** misura quanta informazione una coordinata contiene sul beam;
- l’**Eta-quadro** misura quanto i beam risultano separati lungo quella coordinata.

Se entrambe le metriche risultano più alte per la *latitude* rispetto alla *longitude*, allora esiste una motivazione statistica concreta per cui la rete neurale riesce a imparare meglio usando la latitude.

````python
from sklearn.feature_selection import mutual_info_classif

# Full Scenario-1 dataset — no train/test concern for a purely descriptive analysis.
X_full = df[["latitude", "longitude"]].to_numpy()
y_full = df["beam_index"].to_numpy(dtype=int)
feature_names = ["Latitude", "Longitude"]

# --- Mutual Information ---
mi_scores = mutual_info_classif(X_full, y_full, random_state=42)

# --- Eta-squared (η²) via one-way ANOVA decomposition ---
def eta_squared(col, labels):
    grand_mean = col.mean()
    groups = [col[labels == lbl] for lbl in np.unique(labels)]
    ss_between = sum(len(g) * (g.mean() - grand_mean) ** 2 for g in groups)
    ss_total = ((col - grand_mean) ** 2).sum()
    return ss_between / ss_total

eta2_scores = np.array([eta_squared(X_full[:, i], y_full) for i in range(2)])

# --- Plot ---
fig, axes = plt.subplots(1, 2, figsize=(11, 4))

for ax, scores, title, color, edge in zip(
    axes,
    [mi_scores, eta2_scores],
    ["Mutual Information  I(feature ; beam)", "Eta-quadro  η²  (ANOVA)"],
    ["#2563eb", "#16a34a"],
    ["#1d4ed8", "#15803d"],
):
    bars = ax.bar(feature_names, scores, color=color, alpha=0.82,
                  edgecolor=edge, linewidth=1.5, width=0.45)
    for bar, val in zip(bars, scores):
        ax.text(bar.get_x() + bar.get_width() / 2,
                bar.get_height() + max(scores) * 0.03,
                f"{val:.3f}", ha="center", va="bottom", fontsize=12, fontweight="bold")
    ax.set_title(title, fontsize=11)
    ax.set_ylabel("Score")
    ax.set_ylim(0, max(scores) * 1.3)
    ax.grid(axis="y", alpha=0.3)
    ax.spines[["top", "right"]].set_visible(False)

plt.suptitle("Importanza statistica delle feature GPS (Scenario 1, tutti i campioni)", fontsize=12)
plt.tight_layout()
plt.show()

print(f"Mutual Information — Latitude: {mi_scores[0]:.3f}  |  Longitude: {mi_scores[1]:.3f}")
print(f"Eta-quadro  (η²)  — Latitude: {eta2_scores[0]:.3f}  |  Longitude: {eta2_scores[1]:.3f}")
````

### Separazione visiva: media ± deviazione standard per classe beam

I numeri sopra ci dicono *quanto* latitude è più informativa. Il grafico qui sotto mostra *perché*.

Per ciascuno dei 61 beam presenti nello Scenario 1 calcoliamo la **media** e la **deviazione standard** di latitude e longitude. Ordiniamo i beam per latitudine media crescente e tracciamo entrambe le feature nello stesso ordine.

- Se una feature **separa bene** i beam, ci aspettiamo una curva **monotona e liscia**: beam diversi hanno valori medi diversi e ben distanziati.
- Se una feature **separa male**, la curva sarà **piatta o irregolare**: i beam si mescolano lungo quella coordinata.

````python
# Per ogni beam: media e std di latitude e longitude.
beam_stats = (
    df.groupby("beam_index")
    .agg(
        mean_lat=("latitude",  "mean"),
        std_lat =("latitude",  "std"),
        mean_lon=("longitude", "mean"),
        std_lon =("longitude", "std"),
        n       =("latitude",  "count"),
    )
    .sort_values("mean_lat")          # ordine crescente di latitudine media
    .reset_index()
)

x = np.arange(len(beam_stats))       # posizioni sull'asse x (un punto per beam)

fig, axes = plt.subplots(2, 1, figsize=(14, 7), sharex=True)

# --- Latitude ---
ax = axes[0]
ax.fill_between(x,
                beam_stats["mean_lat"] - beam_stats["std_lat"],
                beam_stats["mean_lat"] + beam_stats["std_lat"],
                color="#93c5fd", alpha=0.35, label="±1 std")
ax.plot(x, beam_stats["mean_lat"], color="#2563eb", lw=2, marker="o",
        ms=4, label="Media per beam")
ax.set_ylabel("Latitude (gradi)")
ax.set_title("Latitude — curva monotona: ogni beam occupa una fascia distinta")
ax.legend(fontsize=9)
ax.grid(alpha=0.25)
ax.spines[["top", "right"]].set_visible(False)

# --- Longitude (stesso ordine sull'asse x) ---
ax = axes[1]
ax.fill_between(x,
                beam_stats["mean_lon"] - beam_stats["std_lon"],
                beam_stats["mean_lon"] + beam_stats["std_lon"],
                color="#fdba74", alpha=0.35, label="±1 std")
ax.plot(x, beam_stats["mean_lon"], color="#ea580c", lw=2, marker="o",
        ms=4, label="Media per beam")
ax.set_ylabel("Longitude (gradi)")
ax.set_title("Longitude — curva irregolare: beam diversi si sovrappongono lungo questa coordinata")
ax.set_xlabel("Beam (ordinati per latitudine media crescente →)")
ax.legend(fontsize=9)
ax.grid(alpha=0.25)
ax.spines[["top", "right"]].set_visible(False)

plt.suptitle("Separazione dei beam lungo latitude vs longitude", fontsize=13, y=1.01)
plt.tight_layout()
plt.show()

# Range delle medie come misura sintetica di separazione
lat_range = beam_stats["mean_lat"].max() - beam_stats["mean_lat"].min()
lon_range = beam_stats["mean_lon"].max() - beam_stats["mean_lon"].min()
print(f"Range delle medie per beam — Latitude: {lat_range:.5f}°  |  Longitude: {lon_range:.5f}°")
print(f"Latitude span / Longitude span = {lat_range / lon_range:.2f}x")
````
